```python
import pandas as pd

fp = r"F:\00.博后研究\城乡连续体制图\New_NEW\提取乡_镇或街道或街道办事处或办事处.csv"
df = pd.read_csv(fp, encoding='utf-8')

def classify_rural(name):
    if pd.isna(name):
        return None
    name = str(name).strip()

    # ==== 1. 农牧型（Rural = 0） ====
    rural_keywords = [
        "乡", "苏木", "农场", "牧场", "垦殖场", "种畜场", "良种繁育中心", "繁育场",
        "养殖场", "渔场", "茶场", "园艺场", "林场", "芦苇场", "农业综合开发区",
        "畜牧场", "农垦", "生态园", "生态区", "农牧场", "牧业", "农牧"
    ]
    if any(k in name for k in rural_keywords):
        return 0

    # ==== 2. 城镇型（Rural = 1） ====
    urban_keywords = [
        "镇", "街道", "街道办", "街道办事处", "街道筹备处", "镇级", "特区", "开发区",
        "经济区", "产业园", "工业园", "园区", "管理区", "管委会", "管理委员会",
        "新区", "社区", "指挥部", "企业集团", "办事处"
    ]
    if any(k in name for k in urban_keywords):
        return 1

    # ==== 3. 特殊判定 ====
    if "公司" in name and "牧" not in name and "农" not in name:
        return 1  # 公司化工业/城市功能
    if "生态" in name and "管理" in name:
        return 1  # 生态管理区，多为开发区性质

    # 默认空值
    return None

df["Rural"] = df["乡镇级"].apply(classify_rural)

out_fp = r"F:\00.博后研究\城乡连续体制图\New_NEW\提取乡_镇街道分类_withRural_v3.csv"
df.to_csv(out_fp, index=False, encoding='utf-8-sig')

print(f"已导出分类结果：{out_fp}")
```
